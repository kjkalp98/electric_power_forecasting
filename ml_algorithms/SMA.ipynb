```python
import csv

p = 5
dict={}
dates = ['25-11-2017.csv','26-11-2017.csv','27-11-2017.csv','28-11-2017.csv','29-11-2017.csv','30-11-2017.csv']
for i in range(len(dates)):
    dict[dates[i]] = []
    time = []
    for d in csv.DictReader(open(dates[i]), delimiter=','):
        time.append(d['time'])
        dict[dates[i]].append(float(d['value']))
n = len(dict[dates[len(dates)-1]])
#actualload = [load[i] for i in range(len(load))]
forecast =[0]*n
for i in range(n):
    for j in range(p):
        forecast[i] += dict[dates[j]][i]/p 
for i in range(n):
    print(time[i]," : ",dict[dates[len(dates)-1]][i]," : ",forecast[i])
```

```python
plt.plot(dict[dates[len(dates)-1]])
plt.plot(forecast)
plt.legend(['load', 'forecast'])
plt.show()
```
